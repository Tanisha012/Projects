## 4. Data Preprocessing

```python
# Drop the 'name' column
df.drop('name', axis=1, inplace=True)

# Separate features and target
X = df.drop('status', axis=1)
Y = df['status']

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=40)

print("Training Set Shape:", X_train.shape, Y_train.shape)
print("Test Set Shape:", X_test.shape, Y_test.shape)
```

## 5. Building and Evaluating Machine Learning Models

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train the Logistic Regression model
log_reg = LogisticRegression().fit(X_train, Y_train)

# Predictions and evaluations
train_preds = log_reg.predict(X_train)
test_preds = log_reg.predict(X_test)

# Train and test accuracy
print("Logistic Regression Model Accuracy on Train:", accuracy_score(Y_train, train_preds))
print("Logistic Regression Model Accuracy on Test:", accuracy_score(Y_test, test_preds))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Train the Random Forest Classifier
RF = RandomForestClassifier().fit(X_train, Y_train)

# Predictions and evaluations
train_preds2 = RF.predict(X_train)
test_preds2 = RF.predict(X_test)

# Train and test accuracy
print("Random Forest Model Accuracy on Train:", accuracy_score(Y_train, train_preds2))
print("Random Forest Model Accuracy on Test:", accuracy_score(Y_test, test_preds2))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds2))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds2))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds2))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds2))

#Wrong Predictions made
print((Y_test !=test_preds2).sum(),'/',((Y_test == test_preds2).sum()+(Y_test != test_preds2).sum()))

# Kappa Score
from sklearn import metrics
print("Kappa Score:", metrics.cohen_kappa_score(Y_test, test_preds2))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Train the Decision Tree Classifier
DT = DecisionTreeClassifier().fit(X_train, Y_train)

# Predictions and evaluations
train_preds3 = DT.predict(X_train)
test_preds3 = DT.predict(X_test)

# Train and test accuracy
print("Decision Tree Model Accuracy on Train:", accuracy_score(Y_train, train_preds3))
print("Decision Tree Model Accuracy on Test:", accuracy_score(Y_test, test_preds3))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds3))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds3))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds3))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds3))

# Wrong Predictions made.
print((Y_test !=test_preds3).sum(),'/',((Y_test == test_preds3).sum()+(Y_test != test_preds3).sum()))
print('-'*50)

# Kappa Score
print("Kappa Score:", metrics.cohen_kappa_score(Y_test, test_preds3))
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

# Train the Naive Bayes Classifier
NB = GaussianNB().fit(X_train, Y_train)

# Predictions and evaluations
train_preds4 = NB.predict(X_train)
test_preds4 = NB.predict(X_test)

# Train and test accuracy
print("Naive Bayes Model Accuracy on Train:", accuracy_score(Y_train, train_preds4))
print("Naive Bayes Model Accuracy on Test:", accuracy_score(Y_test, test_preds4))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds4))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds4))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds4))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds4))
  
# Wrong Predictions made.
print((Y_test !=test_preds4).sum(),'/',((Y_test == test_preds4).sum()+(Y_test != test_preds4).sum()))
print('-'*50)

# Kappa Score
print('KappaScore is: ', metrics.cohen_kappa_score(Y_test,test_preds4))
```

### K-Nearest Neighbors (KNN) Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

# Train the KNN Classifier
KNN = KNeighborsClassifier(weights='distance').fit(X_train, Y_train)

# Predictions and evaluations
train_preds5 = KNN.predict(X_train)
test_preds5 = KNN.predict(X_test)

# Train and test accuracy
print("KNN Model Accuracy on Train:", accuracy_score(Y_train, train_preds5))
print("KNN Model Accuracy on Test:", accuracy_score(Y_test, test_preds5))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds5))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds5))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds5))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds5))

# Wrong Predictions made.
print((Y_test !=test_preds5).sum(),'/',((Y_test == test_preds5).sum()+(Y_test != test_preds5).sum()))
print('-'*50)

# Kappa Score
print("Kappa Score:", metrics.cohen_kappa_score(Y_test, test_preds5))
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

# Train the SVM
SVM = SVC(kernel='linear').fit(X_train, Y_train)

# Predictions and evaluations
train_preds6 = SVM.predict(X_train)
test_preds6 = SVM.predict(X_test)

# Train and test accuracy
print("SVM Model Accuracy on Train:", accuracy_score(Y_train, train_preds6))
print("SVM Model Accuracy on Test:", accuracy_score(Y_test, test_preds6))

# Confusion matrix and classification report
print("Confusion Matrix (Train):\n", confusion_matrix(Y_train, train_preds6))
print("Confusion Matrix (Test):\n", confusion_matrix(Y_test, test_preds6))
print("Classification Report (Train):\n", classification_report(Y_train, train_preds6))
print("Classification Report (Test):\n", classification_report(Y_test, test_preds6))

# Wrong Predictions made.
print((Y_test !=test_preds6).sum(),'/',((Y_test == test_preds6).sum()+(Y_test != test_preds6).sum()))
print('-'*50)
# Kappa Score
print('KappaScore is: ', metrics.cohen_kappa_score(Y_test,test_preds6))
```
